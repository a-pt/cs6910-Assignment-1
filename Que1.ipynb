```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```
from keras.datasets import fashion_mnist
```

```
(train_images,train_labels), (test_images,test_labels)= fashion_mnist.load_data()
```

```
train_images.shape, train_labels.shape, test_images.shape, test_labels.shape
```

```
def get_fashion_mnist_labels(ind):
  label_val = ['Tshirt/Top', 'Trousers', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneakers', 'Bag', 'Ankle Boot']
  return label_val[ind]
```

```
fig = plt.figure(figsize=(20,20))
num_cols = 8
num_rows = 8
for i in range(1, num_cols*num_rows +1):
  ax = fig.add_subplot(num_rows,num_cols,i)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(train_images[i], cmap="gray_r")
  ax.title.set_text(get_fashion_mnist_labels(train_labels[i]))
```
