```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```
from keras.datasets import fashion_mnist
```

```
data_fashion_mnist = fashion_mnist
```

```
data_fashion_mnist
```

```
(train_images,train_labels), (test_images,test_labels)= data_fashion_mnist.load_data()
```

```
train_images.shape, train_labels.shape, test_images.shape, test_labels.shape
```

```
def get_fashion_mnist_labels(ind):
  label_val = ['Tshirt/Top', 'Trousers', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneakers', 'Bag', 'Ankle Boot']
  return label_val[ind]
```

```
plt.figure(figsize=(28,28))
for i in range(32):
  num_cols = 8
  num_rows = 4
  plt.subplot(num_rows,num_cols,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(train_images[i], cmap="gray_r")
  plt.xlabel(get_fashion_mnist_labels(train_labels[i]))
```
